<h1> Imports </h1>

```python
import pandas as pd
import numpy as np
```

#Reading the CSV

```python
dfTest = pd.read_csv('train.csv')
dfTrain = pd.read_csv('test.csv')
```

```python
dfTrain.head()
```

#appending survived coloumn to the test dataset

```python
trainY = pd.read_csv('gender_submission.csv')
```

```python
dfTrain['Survived'] = trainY['Survived']
```

```python
dfTrain.head()
```

```python
df = dfTest.append(dfTrain)
```

```python
df.head()
```

```python
len(df)
```

#Describing the dataset

```python
df.describe()
```

#Droping the rows with no lable (They are useless)

```python
df = df.dropna(subset=['Survived'])
```

#Dropping the cabin - Too many nulls

```python
df = df.drop('Cabin', axis= 1)
```

```python
len(df)
```

```python
df.head()
```

#Filling in the null values with mean for numerical features and mode for categorical features

```python
dic = {"Age" : df.Age.mean, "Fare":df.Fare.mean, "Embarked" : 'S'}
```

```python
df = df.fillna(dic)
```

```python
df.describe()
```

#One Hot encoding the embarked coloumn

```python
C = []
Q = []
S = []

for i in df.Embarked:
  if i == 'C':
    C.append(1)
    Q.append(0)
    S.append(0)
  if(i == 'S'):
    C.append(0)
    Q.append(0)
    S.append(1)
  if(i == 'Q'):
    C.append(0)
    Q.append(1)
    S.append(0)

```

```python
df['EmbarkedC'] = C
df['EmbarkedQ'] = Q
df['EmbarkedS'] = S
```

```python
df.head()
```

```python
df.drop("Embarked", axis = 1, inplace = True)
```

```python
df.head()
```

```python
len(df)
```

#Dropping Ticket coloumn - Not much useful

```python
df.drop(["Ticket"], axis = 1, inplace=True)
```

```python
df.head()
```

#String Indexing the sex coloumn

```python
df["Sex"] = df["Sex"].astype('category')
df["Sex"] = df["Sex"].cat.codes
```

```python
df.head()
```

#Rounding of the age to no decimals

```python
df = df.assign(Age = lambda x : round(x,0))
```
